```python
from sklearn.datasets import load_boston
```

```python
lb=load_boston()
x=lb.data
y=lb.target
columns=lb.feature_names
```

```python
import pandas as pd
import numpy as np
```

```python
lb_df=pd.DataFrame(lb.data)
lb_df.columns=columns
lb_df.head()
```

```python
import seaborn as sb
sb.boxplot(x=lb_df['DIS'])
```

```python

import  matplotlib.pyplot as mp
mp.scatter(lb_df['INDUS'],lb_df['TAX'],)
mp.xlabel('non retail',size=15)
mp.ylabel('full property',size=15)
mp.show()
```

```python
mp.subplots(figsize=(16,8))
mp.scatter(lb_df['INDUS'],lb_df['TAX'],)
mp.xlabel('non retail',size=15)
mp.ylabel('full property',size=15)
mp.show()
```
